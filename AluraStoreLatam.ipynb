### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

```python
ingreso_total_tienda= tienda['Precio'].sum()
ingreso_total_tienda2= tienda2['Precio'].sum()
ingreso_total_tienda3= tienda3['Precio'].sum()
ingreso_total_tienda4= tienda4['Precio'].sum()
```

# 3. Calificación promedio de la tienda

# 4. Productos más y menos vendidos

```python
ventas_por_producto_tienda = tienda['Producto'].value_counts()
ventas_por_producto_tienda2 = tienda2['Producto'].value_counts()
ventas_por_producto_tienda3 = tienda3['Producto'].value_counts()
ventas_por_producto_tienda4 = tienda4['Producto'].value_counts()
```

```python
ventas_por_productos= {
    'Tienda 1' : ventas_por_producto_tienda,
    'Tienda 2' : ventas_por_producto_tienda2,
    'Tienda 3' : ventas_por_producto_tienda3,
    'Tienda 4' : ventas_por_producto_tienda4
}
for tiendas, ventas_por_productos in ventas_por_productos.items():
    print(f'Las ventas por productos de {tiendas} son: ')
    print(ventas_por_productos)
```

```python
# Crear un diccionario con las tiendas y sus respectivos DataFrames
tiendas = {'Tienda 1': tienda, 'Tienda 2': tienda2, 'Tienda 3': tienda3, 'Tienda 4': tienda4}

# Diccionarios para almacenar los productos más y menos vendidos por tienda
productos_mas_vendidos_por_tienda = {}
productos_menos_vendidos_por_tienda = {}

# Recorrer cada tienda y su DataFrame correspondiente
for nombre_tienda, df in tiendas.items():
    # Contar las ocurrencias de cada producto y convertir el índice a tipo string
    conteo_productos = df['Producto'].astype(str).value_counts()

    if not conteo_productos.empty:
        # Obtener el número máximo y mínimo de ventas
        max_ventas = conteo_productos.iloc[0]
        min_ventas = conteo_productos.iloc[-1]

        # Obtener todos los productos con el número máximo de ventas
        productos_mas_vendidos_con_count = {
            producto: max_ventas
            for producto in conteo_productos[conteo_productos == max_ventas].index.tolist()
        }
        productos_mas_vendidos_por_tienda[nombre_tienda] = productos_mas_vendidos_con_count

        # Obtener todos los productos con el número mínimo de ventas
        productos_menos_vendidos_con_count = {
            producto: min_ventas
            for producto in conteo_productos[conteo_productos == min_ventas].index.tolist()
        }
        productos_menos_vendidos_por_tienda[nombre_tienda] = productos_menos_vendidos_con_count
    else:
        # Si el DataFrame está vacío, registrar que no hay datos
        productos_mas_vendidos_por_tienda[nombre_tienda] = {"No hay datos de productos": 0}
        productos_menos_vendidos_por_tienda[nombre_tienda] = {"No hay datos de productos": 0}

# Imprimir los productos más vendidos por tienda
print("Productos más vendidos por tienda:")
for nombre_tienda, productos_con_count in productos_mas_vendidos_por_tienda.items():
    print(f"\n{nombre_tienda}")
    for producto, count in productos_con_count.items():
        print(f"{producto}: {count} ventas")

# Imprimir los productos menos vendidos por tienda
print("\nProductos menos vendidos por tienda:")
for nombre_tienda, productos_con_count in productos_menos_vendidos_por_tienda.items():
    print(f"\n{nombre_tienda}")
    for producto, count in productos_con_count.items():
        print(f"{producto}: {count} ventas")
```

```python
print("Costo de envío promedio por tienda:")
for nombre_tienda, costo_promedio in costo_envio_promedio_por_tienda.items():
    print(f"{nombre_tienda}: {costo_promedio:.2f}")
```

##Listado de las ciudades en las que venden cada una de las tiendas.

```python
# Listar las ciudades (lugares de compra) por tienda
for nombre_tienda, df in tiendas.items():
    ciudades = df['Lugar de Compra'].unique()
    print(f"\n{nombre_tienda}")
    print(f"Total de ciudades: {len(ciudades)}")
    print(sorted(ciudades))
```

##¿Cuál tienda dominda por ciudad?

```python
# Calculamos el valor total vendido por tienda y ciudad
ventas = []

for nombre_tienda, df in tiendas.items():
    # Agrupar por ciudad y sumar el precio total
    ventas_ciudad = df.groupby('Lugar de Compra')['Precio'].sum().reset_index()
    ventas_ciudad['Tienda'] = nombre_tienda
    ventas.append(ventas_ciudad)

# Unimos los resultados de todas las tiendas
ventasdf = pd.concat(ventas, ignore_index=True)

# Encontramos la tienda con mayor venta (sin costo de envío) por ciudad
idx_max = ventasdf.groupby('Lugar de Compra')['Precio'].idxmax()
tienda_dominante = ventasdf.loc[idx_max].sort_values('Lugar de Compra')

# Mostramos el resultado final
print("Tienda dominante por ciudad:\n")
print(tienda_dominante[['Lugar de Compra', 'Tienda', 'Precio']])
```

##Gráfico del mapa de Colombia, qué tienda domina en cada ciudad.

```python
# Extraemos coordenadas promedio por ciudad de los DataFrames originales
coords = pd.concat([
    df[['Lugar de Compra', 'lat', 'lon']] for df in tiendas.values()
], ignore_index=True).groupby('Lugar de Compra')[['lat', 'lon']].mean().reset_index()

# Unimos coordenadas con tienda dominante
map_data = pd.merge(tienda_dominante, coords, on='Lugar de Compra', how='left')
```

```python
import folium
from folium.plugins import MarkerCluster

# Creamos el mapa centrado en Colombia
m = folium.Map(location=[4.5709, -74.2973], zoom_start=6)

# Colores distintos por tienda
colores_tienda = {
    'Tienda 1': 'blue',
    'Tienda 2': 'green',
    'Tienda 3': 'red',
    'Tienda 4': 'orange'
}

# Escalamos la intensidad de color según el valor de ventas
max_venta = map_data['Precio'].max()

# Agrupamos los marcadores para que no se superpongan
cluster = MarkerCluster().add_to(m)

for _, row in map_data.iterrows():
    intensidad = row['Precio'] / max_venta
    color = colores_tienda.get(row['Tienda'], 'gray')
    folium.CircleMarker(
        location=[row['lat'], row['lon']],
        radius=6 + (intensidad * 8),  # tamaño proporcional a ventas
        color=color,
        fill=True,
        fill_opacity=0.6 + (intensidad * 0.4),
        popup=(
            f"<b>Ciudad:</b> {row['Lugar de Compra']}<br>"
            f"<b>Tienda dominante:</b> {row['Tienda']}<br>"
            f"<b>Ventas totales:</b> ${row['Precio']:,.0f}"
        )
    ).add_to(cluster)

# Mostrar el mapa
m
```
